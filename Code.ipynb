```python
from sklearn.model_selection import train_test_split

rest, test = train_test_split(clean_df, test_size=0.1, random_state=4)
train, val = train_test_split(rest, test_size=0.1, random_state=4)
print('training data:', train.shape, 'validation data:', val.shape, 'test data:', test.shape)
```

```python
labels = ['action', 'adventure', 'animation', 'comedy', 'crime', 'drama',
       'fantasy', 'horror', 'mystery', 'romance', 'sci-fi', 'short',
       'thriller']

from tensorflow.keras.preprocessing.image import ImageDataGenerator

data_datagen = ImageDataGenerator(rescale=1./255, horizontal_flip=True)

train_data = data_datagen.flow_from_dataframe(
    train,
    x_col='image_rating',
    y_col=labels,  # name for the labels
    target_size=(224, 224),
    class_mode='raw',
    shuffle = True,
    batch_size=32
)

test_data = data_datagen.flow_from_dataframe(
    test,
    x_col='image_rating',
    y_col=labels,  # name for the labels
    target_size=(224, 224),
    class_mode='raw',
    shuffle = False,
    batch_size=32
)

val_data = data_datagen.flow_from_dataframe(
    val,
    x_col='image_rating',
    y_col=labels,  #name for the labels
    target_size=(224, 224),
    class_mode='raw',
    shuffle = False,
    batch_size=32
)
```

```python
# Visualizing the Train data generator
import matplotlib.pyplot as plt

for i in train_data.next():
    print(len(i))
    plt.imshow(i[3])
    break
```

```python
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Model

# Load the pre-trained EfficientNetB0 model
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Add custom layers on top of the base model
x = GlobalAveragePooling2D()(base_model.output)
x = Dense(512, activation='relu')(x)
predictions = Dense(13, activation='softmax')(x)

# Create the final model
model = Model(inputs=base_model.input, outputs=predictions)

# Freeze the base model's layers
for layer in base_model.layers:
    layer.trainable = False

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Print the model summary
model.summary()
```

```python

# Use the one-hot encoded labels as target data
r = model.fit(
    train_data,
    validation_data=val_data,
    epochs=1,
    steps_per_epoch=len(train_data),
    validation_steps=len(val_data),
    verbose=1
)
```

```python
# plot the loss
plt.plot(r.history['loss'], label='train loss')
plt.plot(r.history['val_loss'], label='val loss')
plt.legend()
plt.show()
plt.savefig('LossVal_loss')

# plot the accuracy
plt.plot(r.history['accuracy'], label='train acc')
plt.plot(r.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
plt.savefig('AccVal_acc_rating_EfficientNetB0')
```

```python
# save it as a h5 file


import tensorflow as tf

# Save the model using the SavedModel format
tf.saved_model.save(model, 'model_EfficientNetB0')
```
